# Polynomial vs Exponential growth
In this exercise, we will explore the differences between the polynomial growth and exponential growth. A **polynomial growth** usually takes the form of a polynomial function of time where the exponents are constant, 
such as $f(n)=2n^2+n+1$ or $f(n)=n^3$. An **exponential growth** is different in that the time variable is in the exponent; for instance, $f(n)=2^n$ and $f(n)=5^n+3$ are both exponential growths. These two type of growths 
exhibit very distinct characteristics, which we will demonstate with a few examples.

## 1) Print the values
Consider a polynomial growth function $f(n)=n^4$ and an exponential growth function $g(n)=2^n$. The following blocks of the code will display a table of the values of the two functions for $n=4,5,6,... ,13$. 

**Question**: Which function will grow faster in the long term?

```python
'''
Todo: Run this cell to import all the necessary packages
To run this cell, click the play button on the left-hand edge of the cell or press ctrl + return
'''

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
'''
Todo: Run this cell to print out the values of the two functions for different values of n
'''

# Display the heading of the table
print("{:<8} {:<20} {:<20}".format('n', 'n^4', '2^n')) 
for n in range(4,21,2):
    # Print the values of n, polynomial growth, and exponential growth
    print("{:<8} {:<20} {:<20}".format(n, n**4, 2**n)) 
```

## 2) Plot the functions
In order to better visualize the trend of the two functions, execute the following block of code to plot them. 

**Question**: Why do we choose to plot the functions for the range $[4,20]$?

```python
'''
Task: Run this cell to plot the two growth functions
'''

# Compute values for plotting 
ns = np.arange(4,21)  # set the range of n to be [4,20]
polynomials = ns**4
exponentials = 2**ns

plt.figure(figsize=(8,6))
plt.rcParams.update({'font.size': 14})
plt.plot(ns, polynomials, label='Polynomial Growth $n^4$')
plt.plot(ns, exponentials, label='Expoenetial Growth $2^n$')
plt.legend()
plt.xlabel('$n$')
plt.show()
```

### Early stages of the spread of a virus

A virus typically will spread exponentially in the beginning if no control measures are implemented, as each infected individual can infect multiple people. The following block of code will plot the growth of the total number of confirmed cases in US in March 2020 in both linear and log scale to better demonstrate the exponential growth behavior. (Exponential functions look like a straight line when plotted in log scale.) The data comes from the COVID-19 Data Repository by Johns Hopkins CSSE on github. 

**Question:** Try to experiment with the code by changing the end date. What happens if we change the end date to a more recent date like 2021-01-01? Include the resulting figure in the write-up. (You can save the figure by right-clicking on the image and select 'Save image as...')

```python
'''
Task: Experiment with the cell by changing the end date. 
'''
############# You can change the end date in the following section ###########################
start = '2020-03-01' # March 1st, 2020
end = '2020-03-31' # March 31st, 2020

# Load the csv from the github repository. If you are having trouble connecting to the internet,
# try saving the csv file in the same folder as the notebook and replace the first line of code with: 
# url='time_series_covid19_confirmed_global.csv'
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
df = pd.read_csv(url)

US_data=df[df['Country/Region'] == 'US'][df.columns[4:]].T
US_data.index=pd.to_datetime(US_data.index)
US_data=US_data.rename(columns={US_data.columns[0]: "US total confirmed cases"})

# Plot the data in linear and log scale
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,5))
f.suptitle('Total COVID cases in United State in March 2020')
US_data[start:end].plot(style='-o',ax=ax1,title='Linear Scale')
US_data[start:end].plot(style='-o',logy=True,ax=ax2,title='Log Scale')
plt.show()
```
